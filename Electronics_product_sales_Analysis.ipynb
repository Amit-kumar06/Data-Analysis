```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# importing dataset(i.e. Electronic_sales)

Electronic_sales=pd.read_csv('Electronic_sales.csv')
print(Electronic_sales)
```

```python
#listof first fiverows
print(Electronic_sales.head(5))
```

```python
#listof last fiverows
print(Electronic_sales.tail(5))
```

```python
# Shape of the dataset to show how much row and columns it has
print(Electronic_sales.shape)
```

```python
# It is also a good practice to know the columns and their corresponding data types. 
# along with finding whether they contain null values or not.

print(Electronic_sales.info())
```

```python
# to get a better understanding of the dataset, 
# wecan also see the statistical summary of the dataset.For example--> Rating

print(Electronic_sales['Rating'].describe())   # Also to show statistical information
```

```python
# We can also see the number of unique users and items in the dataset.

print(Electronic_sales.nunique())
```

```python
# (1)  SALES ANALYSIS BY PRODUCT TYPE.

# Total and average revenue by Product Type

product_sales = Electronic_sales.groupby('Product Type').agg(
    total_revenue=('Total Price', np.sum),
    average_revenue=('Total Price', np.mean),
    total_sales=('Quantity', np.sum)
).sort_values(by='total_revenue', ascending=False)

print(product_sales)

# Plotting total revenue by product type

plt.figure(figsize=(8,6))
plt.bar(product_sales.index, product_sales['total_revenue'], color='red',width=0.6)
plt.title('Total Revenue by Product Type')
plt.xlabel('Product Type')
plt.xticks(rotation=45)
plt.ylabel('Total Revenue')
plt.show()
```

```python
# (2) CUSTOMER DEMOGRAPHICS

# Age distribution

plt.figure(figsize=(10,5))
plt.hist(Electronic_sales['Age'].dropna(), bins=20, color="blue", edgecolor="black",width=1.8)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('No.of Users')
plt.show()

# Gender distribution

plt.figure(figsize=(6,6))
gender_counts = Electronic_sales['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=["orange", "pink"])
plt.title('Gender Distribution')
plt.show()

# Loyalty membership distribution

plt.figure(figsize=(6,6))
loyalty_counts = Electronic_sales['Loyalty Member'].value_counts()
plt.pie(loyalty_counts, labels=loyalty_counts.index, autopct='%1.1f%%', colors=["red", "green"])
plt.title('Loyalty Membership Distribution')
plt.show()
```

```python
# (3) ORDER STATUS

# Order status by product type

order_status_product = Electronic_sales.pivot_table(index='Product Type', columns='Order Status', aggfunc='size', fill_value=0)
order_status_product.plot(kind='bar', stacked=True, figsize=(8,6), color=['red', 'blue'])
print(order_status_product)
plt.title('Order Status by Product Type')
plt.xlabel('Product Type')
plt.xticks(rotation=45)
plt.ylabel('Number of Orders')
plt.show()
```

```python
# (4) PAYMENT METHOD PREFERENCES

# Payment method distribution

plt.figure(figsize=(8,6))
payment_counts = Electronic_sales['Payment Method'].value_counts()
print(payment_counts)
plt.bar(payment_counts.index, payment_counts, color="lightcoral")
plt.title('Payment Method Preferences')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# (5) PURCHASE TRENDS OVER TIME

# Convert Purchase Date to datetime

Electronic_sales['Purchase Date'] = pd.to_datetime(Electronic_sales['Purchase Date'], format='%d-%m-%Y')

# Grouping by month

monthly_sales = Electronic_sales.groupby(Electronic_sales['Purchase Date'].dt.to_period('M')).agg(total_sales=('Total Price', 'sum')).reset_index()
monthly_sales['Purchase Date'] = monthly_sales['Purchase Date'].dt.to_timestamp()
print(monthly_sales)

# Plot monthly sales trends

plt.figure(figsize=(10,6))
plt.plot(monthly_sales['Purchase Date'], monthly_sales['total_sales'], marker='o', color='blue')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# (6) ADD-on PURCJASE ANALYSIS

# Count of each Add-on type

add_ons = Electronic_sales['Add-ons Purchased'].dropna().str.split(',').explode()
add_on_counts = add_ons.value_counts()
print(add_on_counts)

# Plot add-on popularity

plt.figure(figsize=(10, 6))
plt.bar(add_on_counts.index, add_on_counts, color='brown')
plt.title('Most Popular Add-ons')
plt.xlabel('Add-on Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# (7) CUSTOMER RATINGS ANALYSIS

# Average rating by product type

rating_by_product = Electronic_sales.groupby('Product Type')['Rating'].mean().sort_values(ascending=False)
print(rating_by_product)

# Plotting average rating by product type

plt.figure(figsize=(8,6))
plt.bar(rating_by_product.index, rating_by_product, color="plum")
plt.title('Average Rating by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# (8) SHIPPING PREFERENCES

# Shipping type distribution

plt.figure(figsize=(8,6))
shipping_counts = Electronic_sales['Shipping Type'].value_counts()
plt.bar(shipping_counts.index, shipping_counts, color="red")
plt.title('Shipping Type Preferences')
plt.xlabel('Shipping Type')
plt.ylabel('Count')
plt.show()
```

```python
# (9) PRICE AND QUALITY ANALYSIS

# Distribution of Total Price

plt.figure(figsize=(10,5))
plt.hist(Electronic_sales['Total Price'].dropna(), bins=30, color="lightgreen", edgecolor="black")
plt.title('Distribution of Total Price')
plt.xlabel('Total Price')
plt.ylabel('Frequency of Order Prices')
plt.show()

# Relationship between Quantity and Total Price
plt.figure(figsize=(10,6))
plt.scatter(Electronic_sales['Quantity'], Electronic_sales['Total Price'], color='purple', alpha=0.5)
plt.title('Relationship between Quantity and Total Price')
plt.xlabel('Quantity of Items Ordered')
plt.ylabel('Total Price')
plt.show()
```

```python
# (10) TO FIND WHICH PRODUCT HAS HIEGHEST AND LOWEST NUMBER OF SALES


# Find the product with the highest and lowest sales

product_sales = Electronic_sales.groupby('Product Type')['Quantity'].sum()
highest_sales_product = product_sales.idxmax()
lowest_sales_product = product_sales.idxmin()
highest_sales_quantity = product_sales.max()
lowest_sales_quantity = product_sales.min()

print(f"Product with highest sales: {highest_sales_product} ({highest_sales_quantity} units)")
print(f"Product with lowest sales: {lowest_sales_product} ({lowest_sales_quantity} units)")

# Plotting all products
product_sales = Electronic_sales.groupby('Product Type')['Quantity'].sum().sort_values()
plt.figure(figsize=(8,6))
plt.bar(product_sales.index, product_sales.values, color='skyblue')

# Highlight the highest and lowest sales products

plt.bar(product_sales.idxmax(), product_sales.max(), color='green', label='Highest Sales')
plt.bar(product_sales.idxmin(), product_sales.min(), color='red', label='Lowest Sales')

# Adding titles and labels

plt.title('Total Sales by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
